```python
import tensorflow as tf
import numpy as np
import pandas as pd
import pickle
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.boston_housing.load_data()
```

```python
def build_model(n_hidden = 1, n_neurons = 25, optimizer = "SGD", learning_rate = 0.00001, momentum = 0):
    model = keras.models.Sequential()
    model.add(tf.keras.layers.Flatten(input_shape=X_train.shape[1:]))
    for i in range(0, n_hidden):
        model.add(tf.keras.layers.Dense(n_neurons, activation='relu'))
    model.compile(loss="mean_squared_error", optimizer=tf.keras.optimizers.SGD(learning_rate = learning_rate, momentum = momentum, nesterov = (optimizer == "nesterov"))if optimizer =="nesterov" or "sgd"
                      else tf.keras.optimizers.Adam(lr = learning_rate),
                      metrics ="mean_absolute_error")
    model.add(tf.keras.layers.Dense(1))
    model.build()
    return model
```

```python
import os
root_logdir = os.path.join(os.curdir, "tb_logs")
def get_run_logdir(param, val):
    import time
    ts = int(time.time())
    run_id = str(ts) +'_'+ param+ '_' + str(val)
    return os.path.join(root_logdir, run_id)
```

```python
es = tf.keras.callbacks.EarlyStopping(patience=10,min_delta=1,verbose=1, monitor='mean_absolute_error')
```

```python
def reset():
    tf.keras.backend.clear_session()
    np.random.seed(42)
    tf.random.set_seed(42)
```

```python
reset()
lista = []
val = 10**(-6)
tensorboard_cb = tf.keras.callbacks.TensorBoard(get_run_logdir('lr', val))
model = build_model(learning_rate= val)
history = model.fit(X_train, y_train, epochs=100, callbacks=[tensorboard_cb, es])
prediction = model.predict(X_test)
lista.append((val,  mean_squared_error(y_test, prediction),  mean_absolute_error(y_test, prediction)))

val = 10**(-5)
tensorboard_cb = tf.keras.callbacks.TensorBoard(get_run_logdir('lr', val))
model = build_model(learning_rate= val)
history = model.fit(X_train, y_train, epochs=100, callbacks=[tensorboard_cb, es])
prediction = model.predict(X_test)
lista.append((val, mean_squared_error(y_test, prediction), mean_absolute_error(y_test, prediction)))

val = 10**(-4)
tensorboard_cb = tf.keras.callbacks.TensorBoard(get_run_logdir('lr', val))
model = build_model(learning_rate= val)
history = model.fit(X_train, y_train, epochs=100, callbacks=[tensorboard_cb, es])
prediction = model.predict(X_test)
lista.append((val,  mean_squared_error(y_test, prediction), mean_absolute_error(y_test, prediction)))

pickle.dump(lista, open( 'lr' + '.pkl', "wb" ))
```
